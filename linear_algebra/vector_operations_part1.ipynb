# vector operations
- vectors are nothing but arrays in programming

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,-6,scale_units='xy',angles='xy',scale=1,color='b')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Addition of two vectors

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])
```

```python
sum = vector1+vector2
```

```python
sum
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,5,1,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# Subtraction of two vectors

```python
vector1 = np.asarray([0,0,2,3])
vector2 = np.asarray([0,0,3,-2])
diff =sum = vector1-vector2
diff
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-1,5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

# mul of two vectors

```python
vector1 = np.asarray([0,0,2,3])
mul = 3*vector1
mul
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,6,9,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
vector1 = np.asarray([0,0,2,3])
mul = -3*vector1
mul
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-6,-9,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
