```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y,c='red',alpha=0.3,marker='*') # alpha = intensity
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('scatter plot')
plt.grid()
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
```

```python
plt.plot(x,y,'--g')
```

# categorical data

```python
x=np.array(['a','b','c','d','e'])
```

```python
y=np.random.rand(5)
```

```python
plt.bar(x,y)
plt.grid()
plt.title('bar plot')
```

```python
plt.barh(x,y)
plt.grid()
plt.title('barh plot')
```
